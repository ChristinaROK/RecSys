# 1. getPopularityRank

### `popularity` == number of users watched that movie
* ratings = rating.csv에서 user가 평가한 movie개수
* rankings = rating 수를 내림차순해서 매긴 등급

```python
ratings = defaultdict(int) # value의 default값이 0으로 설정됐기 때문에 반복문에서 +=1 이용하기 쉽다 
```

```python
rankings = defaultdict(int)
```

```python
ratingPath = '../ml-latest-small/ratings.csv'
```

```python
with open(ratingPath, newline = '', encoding = 'ISO-8859-1') as csvfile:
    movieReader = csv.reader(csvfile)
    next(movieReader)
    for row in movieReader: # csv module로 read하는 경우는 iterator로 돌려줘야한다. 
        movieID = int(row[1])
        ratings[movieID] = ratings[movieID] + 1 # dictionary의 기본 value가 0으로 설정되어있다. 
rank = 1
for movieID, ratingCount in sorted(ratings.items(), key = lambda x: x[1], reverse = True): # sort by movie's rating
    rankings[movieID] = rank
    rank +=1 # rating 별로 내림차순 한 것에 rank를 달아준다
    
print(sorted(rankings.items(), key = lambda x : x[1], reverse = True))
        
        
```

```python
# I want to sort dict by value! 
from collections import OrderedDict
sorted_rankings = OrderedDict(sorted(rankings.items(), key = lambda x: x[1], reverse = True))
```
