# 밑에껀 효선쓰...!

```python
def get_popularity(df):
    """Get number of clicks that each item received in the df."""

    mask = df["action_type"] == "interaction item info"  # 구하는 action_type
    df_clicks = df[mask]  
    df_item_clicks = (
        df_clicks
        .groupby("reference")
        .size()
        .reset_index(name="n_interaction_item_info")
        .transform(lambda x: x.astype(int))
    )

    return df_item_clicks
```
