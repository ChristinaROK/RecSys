```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
meta = pd.read_csv('item_metadata.csv')
```

```python
prop = meta.properties
```

```python
# apply to all items
newproplist = [item.replace(" ","").split("|") for item in prop]
# alter nested list format into strings
```

```python
proplist = []
for item in newproplist:
    proplist.append(','.join(item))
    
```

```python
proplist[:10]
```

```python
vec = CountVectorizer()
```

```python
X = vec.fit_transform(proplist)
```

```python
# feature (=vocab) list
vocab = vec.get_feature_names()
```

```python
pd.DataFrame(X.toarray(), columns = vocab) # visualization
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
hotel_sim = cosine_similarity(X)
```

```python
hotel_sim.argsort()
```
