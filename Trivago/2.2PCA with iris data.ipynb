```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
pca1 = PCA(n_components = 1)
X_low = pca1.fit_transform(X)
print(X_low)
```

```python
print(pca1.components_) # eigenvector
print(pca1.mean_) # 평균 벡터? 
```
